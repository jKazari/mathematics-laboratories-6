```python
n = int(input())

gons = ['tri', 'tetra', 'penta', 'hexa', 'hepta', 'octa', 'nona', 'deca']

if 3 <= n < 10:
	print(f'A {n}-sided polygon is called a {gons[n-3]}gon')
else:
	print('Not a 3-10 sided polygon')
```

Zadanie 5

```python
def fizzbuzz(n: int = 100):
    for i in range(1, n + 1):
        s = ('fizz' * (i % 3 == 0)) + ('buzz' * (i % 5 == 0))
        yield s or i

print(*fizzbuzz(100), sep='\n')
```

Zadanie 6

```python
word = input()
word_reverse = ''.join(list(word[::-1]))

if word == word_reverse:
	print(f"'{word}' is a palindrome")
else:
	print(f"'{word}' is not a palindrome")
```
