## Inheritance

Inheritance allows features defined in one class to be inherited and reused in the
definition of another class. 

Syntax:

```python
class BaseClassName:
    <statement-1>
    .
    .
    .
    <statement-N>
    
class DerivedClassName(BaseClassName):
    <statement-1>
    .
    .
    .
    <statement-N>
    
```    


The name `BaseClassName` must be acessible from the scope containing derived class e.g.

* in the same module
* with prepended module name:
```python
# in file base.py

class BaseClassName:
    pass

# in file derived.py

class DerivedClassName(base.BaseClassName):
    pass
```

### Overriding methods

Overriding occurs when a method is defined in a class (for example, `Base`) and
also in one of its subclasses (for example, `Derived`). It means that instances of
`Base` and `Derived` both respond to requests for this method to be run but
each has their own implementation of the method.


```python
class Base:
    
    def foo(self):
        print('foo() from Base')
        
        
class Derived(Base):        
    
    def foo(self):
        """Overrides Base.foo()"""
        print('foo() from Derived')
        
    def boo(self):
        """Calls Base.foo()"""
        super().foo()
        
        
        
B1 = Base()
D1 = Derived()

B1.foo()
D1.foo()

D1.boo()
```

## Multiple inheritance


The syntax for defining multiple inheritance in Python allows multiple superclasses to be listed in the parent class list (defined by the brackets following the class name). Each parent class is separated by a comma. 

```python
class SubClassName(BaseClassName1, BaseClassName2, ... BaseClassNameN):
    <class-body>
```

For example:

```python
class Car:
    """ Car """
    
class Toy:
    """ Toy """

class ToyCar(Toy,Car):
    """ A Toy Car """
```

We can say that the class `ToyCar` inherits all the attributes (data) and methods (behaviour) defined in classes `Car` and `Toy`

````python
class Car:
    """ Car """
    def move(self):
        print('BRRR RRR RR')
    
class Toy:
    """ Toy """
    
    def sing(self):
        print('Hupty dumpty sat on a wall')

class ToyCar(Toy,Car):
    """ A Toy Car """
    
    
C1 = Car()
T1 = Toy()
TC1 = ToyCar()


C1.move()
T1.sing()


TC1.move()
TC1.sing()
````

## The diamond problem


```python
class Car:
    """ Car """
    def move(self):
        print('BRRR RRR RR')
    
class Toy:
    """ Toy """

    def move(self):
        print('Tap Tap Tap')
    
    def sing(self):
        print('Humpty Dumpty sat on a wall')

class ToyCar(Toy,Car):
    """ A Toy Car """    
    
toy_car = ToyCar()    
```

`ToyCar` inherits `move()` from both `Car` and `Toy`. 

The question here is which version of the method `move()` will be run when an instance of the `ToyCar` class is instantiated and we call `toy_car.move()`?

In Python,  a breadth first search is used to find methods defined in parent classes; this means that when the method `move()` is called on `ToyCar`, it would first look in `Toy`; it would then only look in `Car` if it could not find a method `move()` in `Toy`. 


However,

```python
class CarToy(Car, Toy):
    """ A Toy Car """    
       
car_toy = CarToy()    
```

`CarToy` inherits `move()` from both `Car` and `Toy`. 

The question here is which version of the method `move()` will be run when an
instance of the `CarToy` class is instantiated and we call `car_toy.move()`?


#### This shows that the order in which a class inherits from multiple classes is significant In Python.

### Multiple Inheritance Considered Harmful

*At first sight multiple inheritance in Python might appear to be particularly useful; after all it allows you to mix together multiple concepts into a single class very easily and quickly. This is certainly true and it can be a very flexible feature if used with care. However, the word care is used here and should be noted.*

*Multiple inheritance can also be very dangerous and is quiet a contentious topic for programmers and for those designing programming languages. Few things in programming are inherently bad but multiple inheritance can result in a level of complexity (and unexpected behaviour) that can tie developers in knots.*

*Part of the problem highlighted by those protesting against multiple inheritance is down to the increased complexity and ambiguity that can occur with multiple inheritance trees that may interconnect between the different classes. One way to think of this is that if a class inherits from multiple classes, then that class may have the same classes in the class hierarchy multiple times, this can make it hard to determine which version of a method may execute and this may allow bugs to go untouched or indeed introduce expected issues due to different interactions between methods.*

## Iterators

Observe that in Python there is a coincise way of looping over collections

```python

    print(element)

    print(key)
```

Behind the scenes, the for statement calls `iter()` on the container object. The function returns an iterator object that defines the method `__next__()` which accesses elements in the container one at a time. When there are no more elements, `__next__()` raises a `StopIteration` exception which tells the for loop to terminate.

````python
#  it is easy to add iterator behavior to your classes
class Backpack:
   
    def __init__(self):
        self.content = dict()
        self._no = 0
        self._index = 0

    def add(self,item,amount):
        self._no += 1
        
        self.content[self._no] = (item,amount)

    def __iter__(self):
        return self
    
    def __next__(self):
        if self._index == len(self.content):
            self._index = 0
            raise StopIteration
        
        self._index += 1
        
        return self.content[self._index]
````

## Generators

Generators are a simple and powerful tool for creating iterators. They are written like regular functions but use the `yield` statement whenever they want to return data. Each time `next()` is called on it, the generator resumes where it left off (it remembers all the data values and which statement was last executed). 

````python
def reverse(data):
    """reverses an interable
       type data must provide len() and __getitem__()
    """
    for index in range(len(data)-1, -1, -1):
        yield data[index]
        
def stright(data):
    """itarateover interable
       type data must provide len() and __getitem__()
    """
    for index in range(0,len(data)):
        yield data[index]
        
````

````python
for i in stright('Backpack'):
    print(i)
````

````python
for i in reverse('Backpack'):
    print(i)
````

````python
#  more advanced example
class Box:
   
    def __init__(self):
        self.content = dict()
        self._no = 0
        self._index = 0

    def add(self,item,amount):
        self._no += 1
        
        self.content[self._no] = (item,amount)

    def __len__(self):
        """returns numbers of items"""
        return self._no
    
    def __getitem__(self,key):
        """For subscripting, quick-and-dirty"""
        return self.content[key+1]
        
    
    def __str__(self):
        """for easy printing"""
        
        if len(self.content) == 0:
            return 'This Box is empty'
        
        out = ''        
        
        for i in self.content.keys():
            out += "There is " + str( self.content[i][1] ) + ' '\
                    + self.content[i][0] + ' in Box\n'
            
        return out
````

## Generator expressions

Some simple generators can be coded succinctly as expressions using a syntax similar to list comprehensions but with parentheses instead of square brackets. These expressions are designed for situations where the generator is used right away by an enclosing function. Generator expressions are more compact but less versatile than full generator definitions and tend to be more memory friendly than equivalent list comprehensions.


````python
sum(i*i for i in range(10))                 # sum of squares
````

````python
xvec = [10, 20, 30]
yvec = [7, 5, 3]
sum(x*y for x,y in zip(xvec, yvec))         # dot product
````
