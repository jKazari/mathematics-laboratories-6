```python
user_poem = []
lengths = []

for i in range(2):
	linijka = input("Podaj linijkę: ")
	user_poem.append(linijka)
	lengths.append(len(linijka))

max_len = max(lengths)+2

print(
	"/" + "-"*max_len + "\\\n" 
	"|" + " "*max_len + "|\n"
	"| " + user_poem[0] + (max_len - len(user_poem[0]) - 1)*" " + "|\n"
	"| " + user_poem[1] + (max_len - len(user_poem[1]) - 1)*" " + "|\n"
	"|" + " "*max_len + "|\n"
	"\\" + "-"*max_len + "/\n"
)
```
