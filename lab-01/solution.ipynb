Zadanie 2.2

```python
WIDTH = 19
HEIGHT = 6
TEXT = "Python"

BOX_COUNT = 2
GAP = 1

TOTAL_WIDTH = BOX_COUNT * WIDTH + (BOX_COUNT - 1) * GAP

text_row = HEIGHT // 2
text_start = 4
text_end = text_start + len(TEXT)

for y in range(HEIGHT):
    for x in range(TOTAL_WIDTH):
        block_size = WIDTH + GAP
        offset_in_block = x % block_size

        if offset_in_block >= WIDTH:
            print(" ", end="")
            continue

        local_x = offset_in_block

        outer_v = local_x == 0 or local_x == WIDTH - 1
        inner_v = local_x == 2 or local_x == WIDTH - 3
        outer_h = y == 0 or y == HEIGHT - 1
        inner_h = y == 1 or y == HEIGHT - 2

        # corners
        if outer_v and y == 0:
            print(" ", end="")
        elif inner_v and y == 1:
            print(" ", end="")
        elif outer_v and y == HEIGHT - 1:
            print("|", end="")
        elif inner_v and y == HEIGHT - 2:
            print("|", end="")

        # horizontal
        elif outer_h:
            print("_", end="")
        elif inner_h and 2 < local_x < WIDTH - 3:
            print("_", end="")

        # vertical
        elif outer_v:
            print("|", end="")
        elif inner_v:
            print("|", end="")

        # text
        elif y == text_row and text_start <= local_x < text_end:
            print(TEXT[local_x - text_start], end="")

        # rest
        else:
            print(" ", end="")

    print()
```
