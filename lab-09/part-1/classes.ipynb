### Pierwsza ciekawa klasa

Definiowanie metod klasy w Pythonie ma specyficzną składnię:


```python
class A:

    def foo(self,<other args>):
        <statements>

```

Pierwszy argument (konwencja nazewnicza`self`) odnosi się do ,,samej siebie''. Jest to odpowiednik `this` z C++


````python
class Box:
       
    def foo(self):
        print("Hello from Box.foo()")
````

````python
# tworzenie obiektów klasy
# class instantiation

# x is an instance of Box
x = Box()

# wywołanie metody
x.foo()
````

Sposob definiowanie atrybutów (pól,) klasy w Pythonie jest inny niż w C++

1. Class variable : współdzielony przez wszystkie instacje
2. Instance variable : unikalna dla instancji

```python
class A:
    
    counter = 3;            # class variable
    
    def __init__(self):   
         self.data = 0      # instance variable
    
```    


````python
# ALERT!!!! Redefinicja Box
class Box:

    vC = 0
    
    def __init__(self):       
        self.vI = 'A'
      
````

````python
x = Box()
y = Box()
z = Box()

print('vC: ',x.vC,y.vC,z.vC,sep='  ')
print('vI: ',x.vI,y.vI,z.vI,sep='  ')
````

````python
#  modyfikacja współdzielonego atrybutu
Box.vC = 11 

print('vC: ',x.vC,y.vC,z.vC,sep='  ')
print('vI: ',x.vI,y.vI,z.vI,sep='  ')
````

````python
# modyfikacja intance variable

x.vI = 'X'
y.vI = 'Jestem sobie zmienna mała'
z.vI = 'Z'

print('vC: ',x.vC,y.vC,z.vC,sep='  ')
print('vI: ',x.vI,y.vI,z.vI,sep='  ')
````

Instance variables can be set by `__init__` on instantiation.

Arguments given to class instantiation operator are passed to `__init__`

````python

# ALERT!!!! Redefinicja Box
class Box:

    vC = 0
    
    def __init__(self, initial_vI):       
        self.vI =  initial_vI
      
````

````python
x = Box('X')

print(x.vI)
````

### Przeanalizować przykład

````python
class Dog:

    kind = 'canine'         # class variable shared by all instances

    def __init__(self, name):
        self.name = name    # instance variable unique to each instance

d = Dog('Fido')
e = Dog('Buddy')


print(d.kind)                  # shared by all dogs

print(e.kind)                  # shared by all dogs

print(d.name)                  # unique to d

print(e.name)                  # unique to e
````

### Przeanalizować przykład

````python
class Dog:

    tricks = []             # mistaken use of a class variable

    def __init__(self, name):
        self.name = name

    def add_trick(self, trick):
        self.tricks.append(trick)

d = Dog('Fido')
e = Dog('Buddy')
f = Dog('Azor')

d.add_trick('roll over')
e.add_trick('play dead')


print(d.name, d.tricks)

print(e.name, e.tricks)

print(f.name, f.tricks)

````

### Bardziej skomplikowana klasa

````python
class WoodenBox():
    
    colour = 'natural'
    
    def __init__(self,w,h,d):
        
        self.width = w
        self.height = h
        self.depth = d
        self.content = []
        
    def add(self,item):
        self.content .append(item)
        
    def unpack(self):
        self.content.clear()
        
    def add_n(self,item,n):
        for i in range(n):
            self.add(item)
        
    def show(self):
        print(self.content)

    def volume(self):
        return self.width * self.height * self.depth
        
````

````python
x = WoodenBox(.5,5,3)

print('volume: ',x.volume())
print('content: ',x.content)
````

````python
y = WoodenBox(3,4,5)
y.add('a')
y.add_n('b',10)
y.show()
````

````python
y.unpack()
y.show()
````

````python
def box_compare(box_1, box_2):
    if box_1.volume() < box_2.volume():
        print('The first box has a smaller capacity.')
    else:
        print('The second box has a smaller capacity.')
````

````python
box_compare(x,y)
````
