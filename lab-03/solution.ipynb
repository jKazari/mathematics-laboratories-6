```python
def can_form_triangle(a, b, c):
    if a <= 0 or b <= 0 or c <= 0:
        return False
    
    if a + b > c and a + c > b and b + c > a:
        return True
    
    return False


a, b, c = 3.0, 4.0, 5.0

if can_form_triangle(a, b, c):
	print(f'Lengths {a, b, c} can form a triangle.')
else:
	print(f'Lengths {a, b, c} cannot form a triangle.')
```

```python
import random

def generate_password():
    length = random.randint(7, 10)
    password = ''

    for _ in range(length):
        ascii_code = random.randint(33, 126)
        password += chr(ascii_code)

    return password

print('Generated password:', generate_password())
```

```python
def express_volume(amount, unit):
    if unit == 'cup':
        total_teaspoons = amount * 48
    elif unit == 'tablespoon':
        total_teaspoons = amount * 3
    elif unit == 'teaspoon':
        total_teaspoons = amount
    else:
        return 'Invalid unit'

    cups = total_teaspoons // 48
    total_teaspoons %= 48

    tablespoons = total_teaspoons // 3
    teaspoons = total_teaspoons % 3

    result = []

    if cups > 0:
        result.append(f'{cups} cup')
    if tablespoons > 0:
        result.append(f'{tablespoons} tablespoon')
    if teaspoons > 0:
        result.append(f'{teaspoons} teaspoon')

    return ', '.join(result)


amount = 59
unit = 'teaspoon'

print(express_volume(amount, unit))
```
