Zadanie 1

```python
exam = {'Andrew', 'Kirsty', 'Beth', 'Emily', 'Sue'}
project = {'Kirsty', 'Emily', 'Ian', 'Stuart'}

print('exam:', exam)
print('project:', project)

both = exam & project
only_exam = exam - project
only_project = project - exam
either_or_both = exam | project
either_not_both = exam ^ project

print("\ntook both exam and project:", both)
print("only took the exam:", only_exam)
print("only submitted the project:", only_project)
print("took either or both:", either_or_both)
print("took either but not both:", either_not_both)
```

```python
numbers = []

while True:
	value = int(input("Enter an integer (0 to stop): "))
	
	if value == 0:
		break
	
	numbers.append(value)

numbers.sort()

print("\nSorted values:")
for num in numbers:
	print(num)
```

Zadanie 3

```python
numbers = []

while True:
    value = input('Enter a number (blank to stop): ')
    
    if value == '':
        break
    
    numbers.append(float(value))

if not numbers:
    print('No values were entered.')
else:
    avg = sum(numbers) / len(numbers)
    print(f'Average: {avg}\n')

    below = []
    equal = []
    above = []

    for num in numbers:
        if num < avg:
            below.append(num)
        elif num > avg:
            above.append(num)
        else:
            equal.append(num)

    print('Below average values:', below)
    print('Average values:', equal)
    print('Above average values:', above)
```

```python
letter_scores = {
    'A': 1, 'E': 1, 'I': 1, 'L': 1, 'N': 1, 'O': 1, 'R': 1, 'S': 1, 'T': 1, 'U': 1,
    'D': 2, 'G': 2,
    'B': 3, 'C': 3, 'M': 3, 'P': 3,
    'F': 4, 'H': 4, 'V': 4, 'W': 4, 'Y': 4,
    'K': 5,
    'J': 8, 'X': 8,
    'Q': 10, 'Z': 10
}

word = input("Enter a word: ").upper()

score = sum(letter_scores.get(letter, 0) for letter in word)

print(f"Scrabble score: {score}")
```
